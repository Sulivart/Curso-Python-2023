# Automatizando el proceso

```python
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
df = pd.DataFrame(response.json())
df.head()
```

```python
resources_list = ('posts', 'comments', 'albums', 'photos', 'todos', 'users')

for resource in resources_list:
  url = f"https://jsonplaceholder.typicode.com/{resource}"
  response = requests.get(url)
  df = pd.DataFrame(response.json())
```

```python
resources_list = ('posts', 'comments', 'albums', 'photos', 'todos', 'users')
new_list = []

for resource in resources_list:
  url = f"https://jsonplaceholder.typicode.com/{resource}"
  response = requests.get(url)
  df = pd.DataFrame(response.json())
  new_list.append(df)
```

```python
# Tabla correspondiente a users
url = "https://jsonplaceholder.typicode.com/users"
response = requests.get(url)
df_users = pd.DataFrame(response.json())

# Nota que la columna addres, cotiene información empaquetada en un formato json
df_users.head()
```

```python
engine = create_engine('postgresql://postgres:postgres@localhost:5432/placeholder')

# 'posts ya fue cargada en los pasos anteriores
# 'users nos arroja el error can´t type 'dict' porque tiene información empaquetada
resources_list = ('comments', 'albums', 'photos', 'todos')

for resource in resources_list:
  url = f"https://jsonplaceholder.typicode.com/{resource}"
  response = requests.get(url)
  df = pd.DataFrame(response.json())
  # El método to_sql tiene el argumento if exists por default
  # es igual a 'fail' por lo que arroja un error si la tabla ya existe
  # Tambien recibe los valores 'replace' y 'append'
  df.to_sql(resource, engine, index = False, if_exists = 'replace')
  print(f'La fuente {resource} fue cargada exitosamente')
```

```python
# Usar el método apply para aplicar la clase
# pd.Series a cada final, y de este modo desempaquetar la información
new_df = df_users["address"].apply(pd.Series)
new_df.head()
```

```python
new_df_2 = new_df["geo"].apply(pd.Series)
new_df_2.head()
```

```python
# pd.concat nos permite unir dos dataframes, ya sea por filas o por columnas
df_final = pd.concat([new_df, new_df_2], axis="columns")
# Con el método drop estamos eliminado la columna `geo`
df_final = df_final.drop("geo")
```
