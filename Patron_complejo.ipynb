```python
! pip3 install ColabTurtle
from ColabTurtle.Turtle import *
import ColabTurtle

lados = int(input('Ingrese el numero de lados: '))
#longitud = int(input('Ingrese la longitud de los lados: '))
grados = 360 / lados

initializeTurtle(initial_speed=10)
pencolor('white')

for i in range(lados):
  pencolor('blue')
  forward(150)
  left(grados)

  for i in range(lados):
    pencolor('red')
    forward(100)
    left(grados)

    for i in range(lados):
      pencolor('green')
      forward(50)
      left(grados)

      for i in range(lados):
        pencolor('yellow')
        forward(25)
        left(grados)
```
